```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(613)

# Create random data
N = 30
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.rand(N) * 1000  # Scales bubble areas

fig, ax = plt.subplots(figsize=(6, 4))
scatter = ax.scatter(x, y, s=sizes, alpha=0.6, c=sizes, cmap='viridis')
ax.set_title('Sample Bubble Chart')
ax.set_xlabel('X-axis value')
ax.set_ylabel('Y-axis value')
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Bubble size (value)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
